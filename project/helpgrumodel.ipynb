```python
apart1=pd.read_csv('apart1.csv')
apart1
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scale_cols = apart1.columns
scaled_df = scaler.fit_transform(apart1)
scaled_df = pd.DataFrame(scaled_df , columns=scale_cols)
scaled_df
```

```python
feature_cols=apart1.columns
label_cols = ['합계']
feature_df = pd.DataFrame(scaled_df, columns = feature_cols)
label_df = pd.DataFrame(scaled_df, columns = label_cols)
label_np = label_df.to_numpy()
feature_np = feature_df.to_numpy()
```

```python
feature_np.shape
```

```python
def make_sequence_dataset(feature, label, window_size):
    feature_list = []
    label_list = []
    for i in range(len(feature)-window_size):
        feature_list.append(feature[i:i+window_size])
        label_list.append(label[i+window_size])
    return np.array(feature_list), np.array(label_list)
```

```python
window_size = 40
X,Y=make_sequence_dataset(feature_np, label_np, window_size)
print(X.shape, Y.shape)
```

```python
Y
```

```python
split= -1000
x_train = X[0:split]
y_train = Y[0:split]
x_test = X[split:]
y_test = Y[split:]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, GRU, Dropout
```

```python
model= Sequential()
model.add(GRU(units=50, activation="relu", return_sequences=True, input_shape = (x_train.shape[1], 5)))
model.add(Dropout(0.2))

model.add(GRU(units=60, activation="relu", return_sequences=True))
model.add(Dropout(0.3))

model.add(GRU(units=80, activation="relu", return_sequences=True))
model.add(Dropout(0.4))

model.add(GRU(units=120, activation="relu"))
model.add(Dropout(0.5))

model.add(Dense(units = 1, activation='linear'))
model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping
model.compile(loss='mse', optimizer = 'adam', metrics=['mae'])
early_stop = EarlyStopping(monitor='val_loss', patience=6)
model.fit(x_train, y_train, validation_data=(x_test,y_test), epochs=100, batch_size=16, callbacks=[early_stop])
```
