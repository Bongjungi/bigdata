```python
def make_sequence_dataset(feature, label, window_size):
    feature_list = []
    label_list = []
    for i in range(len(feature)-window_size-return_size+1):
        feature_list.append(feature[i*window_size:i*window_size+window_size])
        l_list=[]
        for j in range(return_size):
            l_list.extend(label[i+window_size+j])
        label_list.append(l_list)
    return np.array(feature_list), np.array(label_list)
```

```python
window_size = 80
X,Y=make_sequence_dataset(feature_np, label_np, window_size, return_size)
print(X.shape, Y.shape)
```

```python
split= -1000
x_train = X[0:split]
y_train = Y[0:split]
x_test = X[split:]
y_test = Y[split:]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, GRU
```

```python
model = Sequential()
model.add(GRU(256, activation='tanh', input_shape=x_train[0].shape))
model.add(Dense(40, activation='linear'))
model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping
model.compile(loss='mse', optimizer = 'adam', metrics=['mae'])
early_stop = EarlyStopping(monitor='val_loss', patience=6)
model.fit(x_train, y_train, validation_data=(x_test,y_test), epochs=100, batch_size=16, callbacks=[early_stop])
```

```python
pred = model.predict(x_test)
plt.figure(figsize=(30,10))
plt.plot(y_test*1239)
plt.plot(pred*1239)
plt.show()
```
