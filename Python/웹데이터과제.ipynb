```python
# url = "http://www.yes24.com/24/category/bestseller"
# target=urllib.request.urlopen(url)
# contents = target.read()
# soup=BeautifulSoup(contents,'html.parser')
url = 'http://www.yes24.com/24/category/bestseller'
yespage = requests.get(url)
soup=BeautifulSoup(yespage.content, 'html.parser')
#bookname=soup.select("div#bestList > ol> li > p:nth-child(3) > a")
bookname=soup.select("div#bestList > ol> li>p>a>img")
bookname = [e.attrs['alt'] for e in bookname if (e.attrs['alt'] not in(['미리보기','eBook','카트에 넣기','리스트에 넣기']))]
writer=soup.select("div#bestList > ol> li> p.aupu")
price = soup.select("p.price > strong")
dic_list = []
print('순위,책이름, 저자및출판사,가격')
for i in range(len(bookname)):
    dic={}
    dic['순위']=i
    dic['책이름']=bookname[i]
    dic['저자및출판사']=writer[i].text
    dic['가격']=price[i].text
    dic_list.append(dic)
    print(i,bookname[i], writer[i].text, price[i].text,sep=', ')
df= pd.DataFrame(dic_list)
df.to_csv('1_yes24.csv',index=False)

title=[t for t in bookname]
author=[t.text for t in writer]
cost=[t.text for t in price]
df=pd.DataFrame({'순위':range(1,1+len(title)),'책이름':title,'저자및출판사':author, '가격':cost})
df
```

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import time
```

```python
with open('2_한글원본.txt','r',encoding='utf-8') as f:
    ftext = f.read()
ready_list=[]
while len(ftext) > 500:
    temp_str = ftext[:500]
    last_space_idx = temp_str.rfind(' ')
    ready_list.append(ftext[:last_space_idx])
    ftext = ftext[last_space_idx:]
ready_list.append(ftext)

dv=webdriver.Chrome()
dv.get('http://www.naver.com')
elem = dv.find_element_by_name('query')
elem.clear()
elem.send_keys('맞춤법검사기')
elem.send_keys(Keys.RETURN)
time.sleep(2)
txtarea = dv.find_element_by_class_name('txt_gray')
results=""
for ready in ready_list:
    txtarea.clear()
    #txtarea.send_keys(Keys.CONTROL,'a')
    txtarea.send_keys(ready)
    button = dv.find_element_by_class_name('btn_check')
    button.click()
    time.sleep(1)
    soup=BeautifulSoup(dv.page_source, 'html.parser')
    result= soup.select_one('p._result_text.stand_txt').text
    results+=result
dv.get('https://translate.kakao.com/')
close_btn = dv.find_element_by_css_selector('a.btn_close')
close_btn.click()
board = dv.find_element_by_id('query')
ready_list=[]
while len(results) > 5000:
    temp_str = results[:5000]
    last_space_idx = temp_str.rfind(' ')
    ready_list.append(results[:last_space_idx])
    results = results[last_space_idx:]
ready_list.append(results)
results2=""
for ready in ready_list:
    board.clear()
    board.send_keys(ready)
    button = dv.find_element_by_id('btnTranslate')
    button.click()
    time.sleep(5)
    soup=BeautifulSoup(dv.page_source, 'html.parser')
    result= soup.select_one('p#result').text
    results2+=result
with open('2_자동화영어번역본.txt','w', encoding='utf-8') as f:
    f.write(results2)
time.sleep(5)
dv.close()
```

```python
def makelist(filename):
    with open(filename+'.txt','r',encoding='utf-8') as f:
        ftext = f.read()
        ready_list=[]
    while len(ftext) > 500:
        temp_str = ftext[:500]
        last_space_idx = temp_str.rfind(' ')
        ready_list.append(ftext[:last_space_idx])
        ftext = ftext[last_space_idx:]
    ready_list.append(ftext)
    return ready_list
```
