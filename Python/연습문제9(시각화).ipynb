```python
def func(x):
    try:
        int(x)
    except Exception:
        try:
            x=x.replace(',','')
            x=x.replace(' ','')
            int(x)
        except Exception:
            return np.nan
    return float(x)
```

```python
df_last['평당분양가격']=df_last['분양가격(㎡)'].map(func)*3.3
```

# 5.

```python
def func2(x):
    if x=='모든면적':
        return '모든면적'
    else:
        return x.split('전용면적')[1].split('이하')[0].replace('초과','~')
```

```python
df_last['전용면적']=df_last['규모구분'].apply(func2)
```

```python
plt.subplots(figsize=(12,20))
plt.subplots_adjust(wspace=0.2, hspace=1)
dfmax=df['평당분양가격'].max()
for i, location in enumerate(df['지역명'].unique()):
    plt.subplot(9,2,i+1)
    g= sns.barplot(data=df[df['지역명']==location], x='연도',y='평당분양가격',hue='월',ci=None)
    g = sns.pointplot(data=df[df['지역명']==location], x='연도',y='평당분양가격',ci=None)
    g.set(ylim = [0,dfmax], title=location)
    plt.legend().remove()
plt.suptitle('연도별 평당분양가격',y=0.93)
plt.legend(loc='right',bbox_to_anchor=(2.05,0.5),ncol=4, title='월')
plt.show()
```
