```python
# UTM-K
proj_UTMK=Proj(init='epsg:5186')
# WGS1984
proj_WGS84=Proj(init='epsg:4326')
```

```python
x1, y1=200669.94310, 550008.21759
x2, y2= transform(proj_UTMK, proj_WGS84,x1,y1)
```

```python
x2, y2 #경도 위도
```

```python
from geopy.geocoders import Nominatim
service = Nominatim(user_agent="myGeocoder")
```

```python
sinho['경도'],sinho['위도']=transform(proj_UTMK, proj_WGS84,sinho['X좌표'],sinho['Y좌표'])
sinho['위경도']=sinho['위도'].astype(str)+', '+sinho['경도'].astype(str)
```

```python
sinho
```

```python
def makejuso(x):
    location=service.reverse(x)
    if 'district' in location.raw['address'].keys():
        print('.',end='')
        return location.raw['address']['district']
    else:
        print('.',end='')
        return location.raw['address']['suburb']
```

```python
wigyoungdo=sinho['위경도']
```

```python
wigyoungdo.map(makejuso)
```

```python
wigyoung=pd.DataFrame(set(wigyoungdo))
```
