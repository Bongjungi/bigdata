```python
#csv를 numpy배열로 읽어오기
dataset=np.loadtxt('data/pima-indians-diabetes.csv', delimiter=',')
dataset
```

```python
dataset.shape
```

```python
# 훈련셋 : 검증셋 = 7 : 3
# 훈련셋
x_train = dataset[:700, :-1]
y_train = dataset[:700,-1]
#시험셋
x_test = dataset[700:,:-1]
y_test = dataset[700:,-1]
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape,
```

```python
y1_train = utils.to_categorical(y_train, 2)
y1_test = utils.to_categorical(y_test, 2)
```

## 1. 모델 구성

```python
model=Sequential()
model.add(Dense(12, input_dim=8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

## 2. 학습과정 설정

```python
model.compile(loss='binary_crossentropy', optimizer= 'adam', metrics=['accuracy'])
```

## 3. 학습시키기

```python
hist = model.fit(x_train, y_train, epochs=1500, batch_size=64, verbose=2)
```

## 4. 모델 평가하기(학습과정보기, 평가)

```python
# 모델의 학습과정 표시하기
fig, loss_ax = plt.subplots(figsize=(10,6))
loss_ax.plot(hist.history['loss'], 'r', label='loss')
acc_ax = loss_ax.twinx() #loss_ax와 x축 공유
acc_ax.plot(hist.history['accuracy'], 'b', label='accuracy')

loss_ax.set_xlabel('epoch')
loss_ax.set_ylabel('loss')
acc_ax.set_ylabel('accuracy')
loss_ax.legend(loc='upper left')
acc_ax.legend(loc='upper right')
plt.show()
```

```python
model.save('model/pima.h5')
```

```python
# 모델 평가하기
score = model.evaluate(x_test, y_test)
```

```python
print('loss : %s, accuracy : %s'%(score[0], score[1]))
```
